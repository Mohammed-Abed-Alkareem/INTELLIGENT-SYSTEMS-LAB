#### Task 2.10

Mean:

Definition: The mean is the arithmetic average of all data points.
Calculation: Sum all the values and divide by the number of values.
Interpretation on the Histogram: The mean is influenced by all values in the dataset, including outliers. On a histogram, the mean is often marked with a vertical line or simply noted in the title. It represents the "balance point" of the distribution.
Effect of Skewness: In a symmetric distribution, the mean lies at the center. In a skewed distribution, the mean is pulled in the direction of the skew.

Median:

Definition: The median is the middle value when all data points are ordered from smallest to largest.
Calculation: Arrange the data in ascending order and find the middle value (or the average of the two middle values if there is an even number of data points).
Interpretation on the Histogram: The median is less influenced by outliers and skewed data than the mean. On a histogram, the median splits the area under the curve into two equal parts.
Effect of Skewness: In a symmetric distribution, the median coincides with the mean. In a skewed distribution, the median is located closer to the bulk of the data and farther from the tail.

Mode:

Definition: The mode is the value that occurs most frequently in the dataset.
Calculation: Identify the most frequently occurring value(s).
Interpretation on the Histogram: The mode corresponds to the highest peak in the histogram.
Effect of Skewness: There can be more than one mode in a dataset (bimodal or multimodal distributions). The mode is not affected by outliers and is only influenced by the frequency of occurrence.
