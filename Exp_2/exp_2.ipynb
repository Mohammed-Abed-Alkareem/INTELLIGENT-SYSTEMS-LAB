## import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## 1. Data Visualization

### 2.1.1 Using Matplotlib 

```python
t = np.arange(0, 1, 0.001)
sig = np.sin(2 * np.pi *2 * t) 

plt.plot(t, sig)
plt.xlabel('Time [s]')
plt.ylabel('Signal amplitude')
plt.grid(True)

plt.show()
```

```python
t = np.arange(0, 2, 0.001)
sig1 = np.sin(2 * np.pi * 2 * t)
sig2 = np.sin(2 * np.pi * 2 * t - np.pi/6)
sig3 = np.sin(2 * np.pi * 2 * t + np.pi/4)

fig, axs = plt.subplots(2,1)
axs[0].plot(t, sig1)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Signal amplitude')
axs[0].grid(True)

axs[1].plot(t, sig2, t, sig3)
axs[1].set_xlim(0, 2)
axs[1].set_xlabel('Time [s]')
axs[1].set_ylabel('Signal amplitude')
axs[1].grid(True)

fig.tight_layout()
plt.show()
```

#### Task 2.1
 plot two curves in the same figure, add markers of specific size, add a legend, and add a figure title.

```python
t = np.arange(0, 2, 0.001)

sig1 = np.sin(2 * np.pi * 2 * t)
sig2 = np.sin(2 * np.pi * 2 * t - np.pi/6)

plt.plot(t, sig1, label='Signal 1', marker='o', markersize=1)
plt.plot(t, sig2, label='Signal 2', marker='x', markersize=1)

plt.title('Sine wave signals')

plt.xlim(0, 2)
plt.xlabel('Time [s]')
plt.ylabel('Signal amplitude')
plt.grid(True)
plt.legend(loc='upper right')




plt.show()
```

#### Task 2.2: 
Create two axes next to each other, and plot a sinwave in one axis and a cosine wave on the other. Add markers, legend, and a title for the figure.

```python
t = np.arange(0, 2, 0.001)

sig1 = np.sin(2 * np.pi * 2 * t)
sig2 = np.cos(2 * np.pi * 2 * t)

fig, axs = plt.subplots(1,2)
axs[0].plot(t, sig1, label='Signal 1', marker='o', markersize=1, color='red')

axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Signal amplitude')
axs[0].grid(True)
axs[0].legend(loc='upper right')

axs[1].plot(t, sig2, label='Signal 2', marker='x', markersize=1, color='blue')

axs[1].set_xlim(0, 2)
axs[1].set_xlabel('Time [s]')
axs[1].set_ylabel('Signal amplitude')
axs[1].grid(True)
axs[1].legend(loc='upper right')

fig.tight_layout()
plt.title('Sine and Cosine wave signals', y=1)
plt.show()
```

### 2.1.2 Using Seaborn

```python
# generating a sinwwave signal
t = np.arange(0, 1, 0.001)
sin = np.sin(2 * np.pi * 2 * t)
cos = np.cos(2 * np.pi * 2 * t)
#Creating a dataframe from the time, sin, and cos curves
df = pd.DataFrame({'time':t, 'sin':sin, 'cos':cos})
# Create a visualization
sns.relplot(data=df,kind="line",x="time",y="sin").set(title='Sinwave')
```
